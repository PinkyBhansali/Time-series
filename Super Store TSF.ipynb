# Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition= seasonal_decompose(y)

plt.figure(figsize=(15,6))

plt.plot(y, label="Original")
plt.legend(loc="best")

plt.figure(figsize=(15,6))
trend = decomposition.trend
plt.show()
plt.plot(trend,label= "Trend")
plt.legend(loc="best")
plt.xticks(rotation = 90)

plt.figure(figsize=(15,6))
seasonal = decomposition.seasonal
plt.show()
plt.plot(seasonal,label= "seasonal")
plt.legend(loc="best")
plt.xticks(rotation = 90)

plt.figure(figsize=(15,6))
residual = decomposition.resid
plt.show()
plt.plot(residual,label= "Residual")
plt.legend(loc="best")
plt.xticks(rotation = 90)

plt.show()
```

```python
# Time Series Forecasting using Sarima Model - as we have trend and seasonality both
# Sarima -p d q

import itertools


p = d = q = range(0,2) # go till 3 or max 5 not beyond that

pdq= list(itertools.product(p,d,q))

seasonal_pdq = [(x[0],x[1],x[2],12) for x in list(itertools.product(p,d,q))]

print("Ex of parameter combinations for seasonal Arima model....")
print("SARIMA{}x{}".format(pdq[1],seasonal_pdq[1]))
print("SARIMA{}x{}".format(pdq[1],seasonal_pdq[2]))
print("SARIMA{}x{}".format(pdq[2],seasonal_pdq[1]))
print("SARIMA{}x{}".format(pdq[2],seasonal_pdq[1]))
```

```python
# Best Parameter selection using gid search

# As sales data is not only dependent on seasonality but also on other factors 
# we will use SarimaX(x stands for external factors)
```

```python
import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARIMA #(Sarima, SarimaX everyting comes under ARIMA)

from pylab import rcParams

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod=sm.tsa.statespace.SARIMAX(y,order= param,seasonal_order = param_seasonal,enforce_stationarity=False,
                                         enforce_invertibility = False)
            results=mod.fit()
            print("SARIMAX{}x{}12-AIC:{}".format(param,param_seasonal,results.aic))
        except:
            continue
            

```

```python
# SARIMAX(1, 1, 1)x(0, 1, 1, 12)12-AIC:281.38730069394524
# SARIMAX(1, 1, 1)x(1, 1, 1, 12)12-AIC:283.3661014375503
# SARIMAX(0, 1, 1)x(0, 1, 1, 12)12-AIC:279.5806233382824
```

```python
# Fitting the SarimaX model
```

```python
mod=sm.tsa.statespace.SARIMAX(y,order= (0,1,1),seasonal_order = (0,1,1,12),enforce_stationarity=False,
                                         enforce_invertibility = False)
results=mod.fit()

print(results.summary())
```

```python
# Validating Forecast
```

```python
pred = results.get_prediction(start=pd.to_datetime("2017-01-01"),dynamic = False)
pred_ci = pred.conf_int()
ax= y["2014":].plot(label= "observed")
pred.predicted_mean.plot(ax=ax, label="One Step Ahead Forecast",alpha=0.7,figsize =(14,7))
ax.fill_between(pred_ci.index,pred_ci.iloc[:,0],pred_ci.iloc[:,1],color="k",alpha =.2)
ax.set_xlabel("Date")
ax.set_ylabel("Furniture Sales")
plt.legend()
plt.show()
```

```python
# Both actual and predicted line are comiing inside the shaded area so we can say our prediction is right

# But still there is error between actual and predicted we will find that using RMSE
```

```python
# Calculating RMSE to check overall error if my model is good or bad
```

```python
y_forecasted = pred.predicted_mean
y_truth =y["2017-01-01":]
mse = ((y_forecasted-y_truth)**2).mean()
mape = ((y_forecasted-y_truth)/(y_truth)).mean()*100
print("RMSE of our Forecast is {}".format(round(np.sqrt(mse),2)))
print("MAPE is {}".format(round(mape),2))
```
